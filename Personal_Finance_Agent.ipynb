```python
import pandas as pd
```

```python
name = input("Enter your name: ")

income = float(input("Enter your monthly income: "))

food = float(input("Enter your food expenses (numeric value only): "))

transport = float(input("Enter your transport expenses (numeric value only): "))

rent = float(input("Enter your rent expenses (numeric value only): "))

others = float(input("Enter other expenses (numeric value only): "))
```

```python
total_expenses = food + transport + rent + others

balance = income - total_expenses

savings = balance

print("Total Expenses =",total_expenses)

print("Remaining Balance =",balance)
```

```python
if total_expenses > income:

    print("\nWARNING!!!")

    print("You spent more than your income.")

elif total_expenses >= (income*0.8):

    print("\nFinancial Advice:")

    print("Your expenses are very high.")

    print("Try reducing unnecessary spending.")

else:

    print("\nExcellent!")

    print("You are managing your finances well.")

    print("Keep saving consistently.")
```

```python
print("\n------PERSONAL FINANCIAL REPORT------")

print("Name:",name)

print("Monthly Income: ₦",income)

print("Food Expenses: ₦",food)

print("Transport Expenses: ₦",transport)

print("Rent Expenses: ₦",rent)

print("Other Expenses: ₦",others)

print("Total Expenses: ₦",total_expenses)

print("Remaining Balance: ₦",balance)
```

```python
import matplotlib.pyplot as plt

expenses = [food,transport,rent,others]

labels = ["Food","Transport","Rent","Others"]

plt.pie(expenses,
        labels=labels,
        autopct="%1.1f%%")

plt.title("Monthly Expenses")

plt.show()
```

```python
if balance <= 0:
    print("\nYou have exceeded your budget this month.")
    print("Consider reducing unnecessary expenses.")

elif balance <= 50000:
    print("\nYour savings are quite low.")
    print("Try setting aside at least 20% of your income.")

elif balance <= 100000:
    print("\nGood job!")
    print("You have managed your finances fairly well.")

else:
    print("\nExcellent financial management!")
    print("You are saving a healthy amount this month.")
    print("Consider investing part of your savings.")
```

```python
import pandas as pd

data = {
    "Name":[name],
    "Income":[income],
    "Total Expenses":[total_expenses],
    "Balance":[balance]
}

df = pd.DataFrame(data)

df.to_csv("financial_report.csv",index=False)

print("Financial report saved successfully!")
```
