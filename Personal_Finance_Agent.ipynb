```python
from pyngrok import ngrok
from google.colab import userdata

# IMPORTANT: Ensure your NGROK_TOKEN is set in Colab's Secrets (the key name should be 'NGROK_TOKEN')
# If you prefer to paste it directly, replace userdata.get("NGROK_TOKEN") with your token string.
NGROK_AUTH_TOKEN = userdata.get("NGROK_TOKEN") # Assuming 'NGROK_TOKEN' is the secret name
ngrok.set_auth_token(NGROK_AUTH_TOKEN)
```

```python
%cd /content/finance-agent
import subprocess, time
from pyngrok import ngrok

# Ensure all previous ngrok processes are killed
ngrok.kill()
!pkill -f streamlit
time.sleep(5) # Increased sleep to allow ngrok to fully clean up

log_file = open("streamlit_log.txt", "w")
process = subprocess.Popen(
    ["streamlit", "run", "app.py", "--server.port", "8502", "--server.headless", "true"], # Changed port to 8502
    stdout=log_file, stderr=subprocess.STDOUT
)

time.sleep(6)

if process.poll() is not None:
    print("❌ Streamlit crashed on startup. Log output:")
    !cat streamlit_log.txt
else:
    print("✅ Streamlit is running, opening tunnel...")
    public_url = ngrok.connect(8502, "http") # Changed port to 8502
    print("Your app is live at:", public_url)
```
